as there are no null values and duplicated values now changing the format of Camp_Start_Date and Camp_End_Date from object to dateTime format

```python
import datetime
Health_Camp_Detail['Camp_Start_Date']=pd.to_datetime(Health_Camp_Detail['Camp_Start_Date'])
Health_Camp_Detail['Camp_End_Date']=pd.to_datetime(Health_Camp_Detail['Camp_End_Date'])
```

```python
Health_Camp_Detail.info()
```

```python
Health_Camp_Detail['Health_Camp_ID'].value_counts()
```

Creating new columns of month quarter and duration of Health Camp

```python
#Camp_Start_Date
Health_Camp_Detail['Camp_Start_Month']=Health_Camp_Detail['Camp_Start_Date'].apply(lambda x:x.month)
Health_Camp_Detail['Camp_Start_Day']=Health_Camp_Detail['Camp_Start_Date'].apply(lambda x:x.day)
Health_Camp_Detail['Camp_Start_Quarter']=Health_Camp_Detail['Camp_Start_Date'].apply(lambda x:x.quarter)

#Camp_End_Date
Health_Camp_Detail['Camp_End_Month']=Health_Camp_Detail['Camp_End_Date'].apply(lambda x:x.month)
Health_Camp_Detail['Camp_End_Day']=Health_Camp_Detail['Camp_End_Date'].apply(lambda x:x.day)
Health_Camp_Detail['Camp_End_Quarter']=Health_Camp_Detail['Camp_End_Date'].apply(lambda x:x.quarter)

#Camp_Duration
Health_Camp_Detail['Camp_Duration']=(Health_Camp_Detail['Camp_End_Date']-Health_Camp_Detail['Camp_Start_Date']).astype('timedelta64[D]')
```

Camp Duration analysis

first we will check for outliers

```python
sns.boxplot(Health_Camp_Detail.Camp_Duration)
```

As we can see there are outliers. We will remove them for better accuracy.

```python
q1 , q3 = Health_Camp_Detail.Camp_Duration.quantile([0.25,0.75])
print(q1 , q3)
```

```python
iqr = q3 - q1
print(iqr)
```

```python
ll = q1 - 1.5*(iqr)
ul = q3 + 1.5*(iqr)
print (ll,ul)
```

```python
Health_Camp_Detail[Health_Camp_Detail.Camp_Duration>ul]
```

```python
len(Health_Camp_Detail[Health_Camp_Detail.Camp_Duration>ul])
```

```python
Health_Camp_Detail.Camp_Duration=np.where(Health_Camp_Detail['Camp_Duration']>ul,ul,Health_Camp_Detail.Camp_Duration)
```

```python
sns.boxplot(Health_Camp_Detail.Camp_Duration)
```

```python
sns.displot(x=Health_Camp_Detail.Camp_Duration , data=Health_Camp_Detail , kde=True)
plt.ylabel('No of camps')
plt.title('Camps Duration Distribution')
```

we can clearly see that duration of most number of camps was between 0 to 25 days.
